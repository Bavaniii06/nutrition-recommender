```python
import pandas as pd

# Load the cleaned data
df = pd.read_csv('C:/Users/Bavaniii/Downloads/health-nutrition-recommender/data/processed/foods_clean.csv')
df.head()
```

```python
def daily_calorie_need(sex, weight_kg, height_cm, age, activity_factor=1.2):
    if sex.lower() in ['male','m']:
        bmr = 88.362 + 13.397*weight_kg + 4.799*height_cm - 5.677*age
    else:
        bmr = 447.593 + 9.247*weight_kg + 3.098*height_cm - 4.330*age
    return bmr * activity_factor

# Example user
calories_needed = daily_calorie_need('female', 55, 160, 25, 1.375)
print("Daily calories needed:", int(calories_needed))
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Create target vector
target = np.array([[calories_needed, 0, 0, 0]])  # Calories, Protein, Fat, Carbs (others 0 for now)

# Create feature matrix from dataset
features = df[['Calories', 'protein_g', 'fat_g', 'carbs_g']].fillna(0).to_numpy()

# Compute cosine similarity
sims = cosine_similarity(target, features)[0]

# Add similarity scores to dataframe
df['similarity'] = sims

# Get top 10 recommended foods
top_recs = df.sort_values('similarity', ascending=False).head(10)
top_recs[['food_name', 'Calories', 'protein_g', 'fat_g', 'carbs_g']]
```
